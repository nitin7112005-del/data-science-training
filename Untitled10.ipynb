Q9. Bivariate Relationship Visualization (Age vs BMI)

```python
print("--- Age aur BMI ki Statistical Summary ---")
print(df[['Age', 'BMI']].describe())
print("\n" + "="*40 + "\n")
correlation = df['Age'].corr(df['BMI'])
print("--- Correlation Analysis ---")
print(f"Age aur BMI ke beech Correlation: {correlation:.4f}")
if correlation > 0.5:
    print("Insight: Age badhne ke saath BMI mein kafi badhotri ho rahi hai (Strong Positive Relationship).")
elif correlation > 0 and correlation <= 0.5:
    print("Insight: Age badhne ke saath BMI thoda badh raha hai (Weak Positive Relationship).")
elif correlation < 0 and correlation >= -0.5:
    print("Insight: Age badhne ke saath BMI thoda kam ho raha hai (Weak Negative Relationship).")
elif correlation < -0.5:
    print("Insight: Age badhne ke saath BMI mein kafi kami ho rahi hai (Strong Negative Relationship).")
else:
    print("Insight: Age aur BMI ke beech koi linear rishta nahi dikh raha hai.")
```

Q10. Distribution Frequency Analysis (Stress)

```python
print("--- Stress Level Frequency Count ---")
stress_counts = df['Stress_Level'].value_counts().sort_index()
print(stress_counts)
print("\n" + "="*40 + "\n")
print("--- Stress Level Percentage Distribution ---")
stress_percentage = df['Stress_Level'].value_counts(normalize=True).sort_index() * 100
for level, pct in stress_percentage.items():
    print(f"Stress Level {level}: {pct:.2f}% log")
print("\n" + "="*40 + "\n")
print("--- Stress Level Statistical Summary ---")
print(df['Stress_Level'].describe())
```
